```python
def generate_N_grams(text, ngram):
    words = [word for word in text.split(" ")]
    temp = zip(*[words[i:] for i in range(0,ngram)])
    return [' '.join(ngram) for ngram in temp]
```

### Get all N-grams that have atleast a sim score

```python
df_1 = uni_df[uni_df['sim']>0]
df_2 = bi_df[bi_df['sim']>0]
df_3 = tri_df[tri_df['sim']>0]
df_4 = quad_df[quad_df['sim']>0]
df_5 = pent_df[pent_df['sim']>0]
```

```python
df = pd.concat([df_1, df_2, df_3, df_4, df_5])
df = df.rename(columns={0: "n-gram", 1: "frequency"})
df = df.sort_values('sim', ascending = False).reset_index(drop=True)
```

```python
df['type of n-gram'] = df['n-gram'].apply(lambda x: len(x.split(" ")))
```

```python
dict_ngrams = {1: 'unigram',
               2: 'bi-gram',
               3: 'tri-gram',
               4: 'quad-gram',
               5: 'pent-gram'}
```

```python
df['type of n-gram'] = df['type of n-gram'].map(dict_ngrams)
```
